```python
#
# Variveis de entrada e inicializao de valores
#
# ni ... nmero de iteraes
# xi ... valor constante a ser adicionado na soma
# S  ... valor final do somatrio
# Si ... acumulador do somatrio
# Si = 0 (inicializao do somatrio)
# 
#
Si=0

ni=int(input("Digite o nmero de iteraes ===> "))
xi=float(input("Digite o valor da constante xi ==> "))

for itera in range(ni):
    Si=Si+xi

print("S = ",Si)
```

# Preciso da Mquina
A preciso da mquina  definida como sendo o menor nmero positivo em aritmtica de ponto flutuante, epsilon.
Este nmero depende totalmente do sistema de representao da mquina: base numrica, total de dgitos na mantissa, da forma como so realizadas as operaes e do computador utilizado.

```python
a = float(1)
while (1 + a) > 1:
    a = a/2
prec = 2*a
print("O menor valor : ",prec)
```
