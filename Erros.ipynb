```python
#
# Variáveis de entrada e inicialização de valores
#
# ni ... número de iterações
# xi ... valor constante a ser adicionado na soma
# S  ... valor final do somatório
# Si ... acumulador do somatório
# Si = 0 (inicialização do somatório)
# 
#
Si=0

ni=int(input("Digite o número de iterações ===> "))
xi=float(input("Digite o valor da constante xi ==> "))

for itera in range(ni):
    Si=Si+xi

print("S = ",Si)
```

# Precisão da Máquina
A precisão da máquina é definida como sendo o menor número positivo em aritmética de ponto flutuante, epsilon.
Este número depende totalmente do sistema de representação da máquina: base numérica, total de dígitos na mantissa, da forma como são realizadas as operações e do computador utilizado.

```python
a = float(1)
while (1 + a) > 1:
    a = a/2
prec = 2*a
print("O menor valor é: ",prec)
```
