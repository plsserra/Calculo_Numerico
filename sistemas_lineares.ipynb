```python
#
# Importação do pacote NumPy
#
import numpy as np
```

```python
#
# Leitura e definição das variáveis
#
def leia_dados():
    m=0
    n=0
    print("\t__________________________________________________________\n")
    n = int(input("\t>>> Digite o número de incógnitas do sistema >>> "))
    m = int(input("\t>>> Digite o número de equações do sistema >>> "))
    print("\n\t__________________________________________________________\n")
    while m != n:
        print("\n\t>>> O número de incógnitas está diferente do número de equações <<<")
        print("\t>>> Isto é uma evidência de que o sistema é Possível e Indeterminado - SPI <<<")
        print("\t>>> ou se trata de um sistema Impossível - SI <<<")
        print("\t>>> Verifique seus dados e digite novamente <<<\n")
        print("\t__________________________________________________________\n")
        n = int(input("\t>>> Digite o número de incógnitas do sistema >>> "))
        m = int(input("\t>>> Digite o número de equações do sistema >>> "))
        print("\n\t__________________________________________________________\n")
    a=np.zeros((m,n),dtype=np.float64)
    b=np.zeros(m,dtype=np.float64)
    for i in range(m):
        for j in range(n):
            print("\n\t>>> Digite o termo A[",i+1,"][",j+1,"]: ",end = " ")
            a[i,j]=float(input(" "))
        print("\n\t>>> Digite o termo b[",i+1,"]",end = " ")
        b[i]=float(input(" "))
    return a,b,m,n
```

```python
def escreva_dados(a,b,m,n):
    print("\n\t____________________________________________")
    print("\t >>> Resolução de Sistemas Lineares")
    print("\t >>> Ax = b")
    print("\t >>> Número de equações: ",m)
    print("\t >>> Número de incógnitas: ",n)
    print("\t____________________________________________")
    print("\t >>> Matriz A: \n")
    for i in range(m):
        print("\n\t",end="")
        for j in range(n):
            print(a[i,j],end="\t")
    print("\n\t____________________________________________")
    print("\t >>> Matriz b: \n\t")
    for i in range(m):
        print("\t",b[i])
```

```python
#
# Resolução de Sistemas Lineares pelo método da Eliminação de Gauss
#
def met_gauss(a,b,m,n):
    #
    # Algoritmo de Eliminação ou Triangularização da matriz A
    #
    # Inicialização do vetor de variáveis "x"
    #
    
    x=np.zeros(n,dtype=np.float64)
    for k in range(n-1):
        for i in range(k+1,n):
            mult=a[i,k]/a[k,k]
            a[i,k]=0
            for j in range(k+1,n):
                a[i,j]=a[i,j]-mult*a[k,j]
            b[i]=b[i]-mult*b[k]
    #
    # Cálculo da resolução do sistema
    #
    x[n-1]=b[n-1]/a[n-1,n-1]
    for k in range(n-2,-1,-1):
        s=0
        for j in range(k,n):
            s=s+a[k,j]*x[j]
        x[k]=(b[k]-s)/a[k,k]
    return x
```

```python
a,b,m,n=leia_dados()
escreva_dados(a,b,m,n)
x=met_gauss(a,b,m,n)
for i in range(n):
    print("\n\t x[",i,"] = %.4f"%x[i])
```
